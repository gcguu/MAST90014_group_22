```python
import pandas as pd
import numpy as np


# Products

products_df = pd.DataFrame({
    "product_id": ["milk_full", "milk_oat", "milk_soy", "milk_almond", "bean_house", "bean_single", "bean_decaf"],
    "product_name": ["Full Cream Milk", "Oat Milk", "Soy Milk", "Almond Milk", "House Beans", "Single Origin", "Decaf Beans"],
    "weight_per_unit_kg": [1.03, 1.03, 1.03, 1.03, 1.00, 1.00, 1.00],
    "margin_per_unit": [2.5, 3.2, 3.0, 3.5, 8.0, 8.5, 7.5]
})


# Cafes

cafes_df = pd.DataFrame({
    "cafe_id": [f"cafe_{i:02d}" for i in range(1, 35)]
})


# Vans

vans_df = pd.DataFrame({
    "van_id": [1, 2],
    "capacity_kg": [500, 700],
    "fuel_cost_per_km": [0.40, 0.50]
})


# Demand

demands_df = pd.DataFrame({
    "cafe_id": ["cafe_01","cafe_01","cafe_01","cafe_01","cafe_01","cafe_01","cafe_01",
        "cafe_02","cafe_02","cafe_02","cafe_02","cafe_02","cafe_02","cafe_02",
        "cafe_03","cafe_03","cafe_03","cafe_03","cafe_03","cafe_03","cafe_03",
        "cafe_04","cafe_04","cafe_04","cafe_04","cafe_04","cafe_04","cafe_04",
        "cafe_05","cafe_05","cafe_05","cafe_05","cafe_05","cafe_05","cafe_05",
        "cafe_06","cafe_06","cafe_06","cafe_06","cafe_06","cafe_06","cafe_06",
        "cafe_07","cafe_07","cafe_07","cafe_07","cafe_07","cafe_07","cafe_07",
        "cafe_08","cafe_08","cafe_08","cafe_08","cafe_08","cafe_08","cafe_08",
        "cafe_09","cafe_09","cafe_09","cafe_09","cafe_09","cafe_09","cafe_09",
        "cafe_10","cafe_10","cafe_10","cafe_10","cafe_10","cafe_10","cafe_10",
        "cafe_11","cafe_11","cafe_11","cafe_11","cafe_11","cafe_11","cafe_11",
        "cafe_12","cafe_12","cafe_12","cafe_12","cafe_12","cafe_12","cafe_12",
        "cafe_13","cafe_13","cafe_13","cafe_13","cafe_13","cafe_13","cafe_13",
        "cafe_14","cafe_14","cafe_14","cafe_14","cafe_14","cafe_14","cafe_14",
        "cafe_15","cafe_15","cafe_15","cafe_15","cafe_15","cafe_15","cafe_15",
        "cafe_16","cafe_16","cafe_16","cafe_16","cafe_16","cafe_16","cafe_16",
        "cafe_17","cafe_17","cafe_17","cafe_17","cafe_17","cafe_17","cafe_17",
        "cafe_18","cafe_18","cafe_18","cafe_18","cafe_18","cafe_18","cafe_18",
        "cafe_19","cafe_19","cafe_19","cafe_19","cafe_19","cafe_19","cafe_19",
        "cafe_20","cafe_20","cafe_20","cafe_20","cafe_20","cafe_20","cafe_20",
        "cafe_21","cafe_21","cafe_21","cafe_21","cafe_21","cafe_21","cafe_21",
        "cafe_22","cafe_22","cafe_22","cafe_22","cafe_22","cafe_22","cafe_22",
        "cafe_23","cafe_23","cafe_23","cafe_23","cafe_23","cafe_23","cafe_23",
        "cafe_24","cafe_24","cafe_24","cafe_24","cafe_24","cafe_24","cafe_24",
        "cafe_25","cafe_25","cafe_25","cafe_25","cafe_25","cafe_25","cafe_25",
        "cafe_26","cafe_26","cafe_26","cafe_26","cafe_26","cafe_26","cafe_26",
        "cafe_27","cafe_27","cafe_27","cafe_27","cafe_27","cafe_27","cafe_27",
        "cafe_28","cafe_28","cafe_28","cafe_28","cafe_28","cafe_28","cafe_28",
        "cafe_29","cafe_29","cafe_29","cafe_29","cafe_29","cafe_29","cafe_29",
        "cafe_30","cafe_30","cafe_30","cafe_30","cafe_30","cafe_30","cafe_30",
        "cafe_31","cafe_31","cafe_31","cafe_31","cafe_31","cafe_31","cafe_31",
        "cafe_32","cafe_32","cafe_32","cafe_32","cafe_32","cafe_32","cafe_32",
        "cafe_33","cafe_33","cafe_33","cafe_33","cafe_33","cafe_33","cafe_33",
        "cafe_34","cafe_34","cafe_34","cafe_34","cafe_34","cafe_34","cafe_34"],

    "product_id": ["milk_full","milk_oat","milk_soy","milk_almond","bean_house","bean_single","bean_decaf"] * 34,

    "daily_demand":
        [11,3,1,2,1,1,1,51,12,7,6,5,3,2,14,6,3,2,2,2,1,23,6,3,1,3,2,1,
        13,3,2,2,2,1,0,16,3,3,2,1,1,1,14,2,2,1,1,1,1,48,14,8,6,3,1,1,
        22,3,5,3,3,1,0,21,5,4,1,3,1,1,16,3,1,2,1,0,1,35,7,4,7,3,3,2,
        31,11,6,4,3,3,1,16,3,3,1,1,1,0,16,3,3,1,1,0,0,11,3,1,1,1,1,0,
        10,2,2,1,1,0,0,24,7,5,1,3,1,0,19,6,3,2,1,2,0,14,3,2,2,1,1,1,
        20,6,2,3,2,1,0,10,4,2,1,2,0,0,12,4,2,2,1,0,1,14,2,3,2,2,1,0,
        24,3,5,1,2,1,1,17,6,3,3,2,1,0,14,2,1,1,1,1,1,25,7,5,1,1,1,1,
        15,7,5,2,2,2,0,13,4,3,2,1,0,0,19,7,5,2,2,2,0,16,2,1,2,1,1,1,
        14,4,3,2,1,1,0,54,8,7,5,3,1,2]
})


# Coordinates

nodes = {
    "depot": (-37.7982158, 144.9636857),

    "cafe_01": (-37.8023383, 144.9851717),
    "cafe_02": (-37.8146934, 144.9582383),
    "cafe_03": (-37.8029711, 144.9590027),
    "cafe_04": (-37.7948113, 144.9781341),
    "cafe_05": (-37.776348, 144.972123),
    "cafe_06": (-37.8312475, 144.960354),
    "cafe_07": (-37.774643, 144.962468),
    "cafe_08": (-37.8226075, 145.0342781),
    "cafe_09": (-37.816791, 144.966162),
    "cafe_10": (-37.7947678, 144.9342535),
    "cafe_11": (-37.7763047, 144.9612147),
    "cafe_12": (-37.8028093, 144.9870052),
    "cafe_13": (-37.813541, 144.9620988),
    "cafe_14": (-37.7959659, 144.9799313),
    "cafe_15": (-37.8043369, 144.9545001),
    "cafe_16": (-37.7641144, 144.9732824),
    "cafe_17": (-37.799384, 144.9781281),
    "cafe_18": (-37.7981641, 144.9858767),
    "cafe_19": (-37.8394531, 144.9963319),
    "cafe_20": (-37.8269479, 144.9929394),
    "cafe_21": (-37.775525, 144.9601403),
    "cafe_22": (-37.8091416, 145.006767),
    "cafe_23": (-37.845757, 144.993288),
    "cafe_24": (-37.8169555, 145.0131324),
    "cafe_25": (-37.8108846, 144.9455841),
    "cafe_26": (-37.7847929, 144.9823376),
    "cafe_27": (-37.761889, 144.961811),
    "cafe_28": (-37.8054501, 144.9542036),
    "cafe_29": (-37.810993, 144.964066),
    "cafe_30": (-37.8158579, 144.9720281),
    "cafe_31": (-37.8014, 144.9672),
    "cafe_32": (-37.8109, 144.9668),
    "cafe_33": (-37.7775, 144.9722),
    "cafe_34": (-37.8175, 144.964)
}

node_ids = list(nodes.keys())


# Using Haversine formula to calculate shortest distance "as the crow flies"

def haversine(lat1, lon1, lat2, lon2):

    R = 6371 # Mean radius of the Earth in KM

    lat1, lon1, lat2, lon2 = map(np.radians, [lat1, lon1, lat2, lon2])

    dlat = lat2 - lat1
    dlon = lon2 - lon1

    a = (np.sin(dlat / 2) ** 2 + np.cos(lat1) * np.cos(lat2) * np.sin(dlon / 2) ** 2)

    return 2 * R * np.arcsin(np.sqrt(a))


# Distance matrix

distance_matrix = pd.DataFrame(
    np.zeros((len(node_ids), len(node_ids))),
    index=node_ids,
    columns=node_ids
)

for i in node_ids:
    for j in node_ids:
        lat1, lon1 = nodes[i]
        lat2, lon2 = nodes[j]

        distance_matrix.loc[i, j] = haversine(lat1, lon1, lat2, lon2)

distance_matrix = distance_matrix.round(3)


# Compute the total weight and total revenue for each cafe

cafe_data = {}

for cafe_id in demands_df["cafe_id"].unique():

    cafe_dem = demands_df[demands_df["cafe_id"] == cafe_id]

    total_weight = 0
    total_revenue = 0

    for _, row in cafe_dem.iterrows():

        product = products_df[products_df["product_id"] == row["product_id"]].iloc[0]

        total_weight += row["daily_demand"] * product["weight_per_unit_kg"]
        total_revenue += row["daily_demand"] * product["margin_per_unit"]

    cafe_data[cafe_id] = {
        "weight": total_weight,
        "revenue": total_revenue
    }
    

# Nearest neighbour heuristic

def nearest_neighbour(distance_matrix, vans_df, cafe_data):

    unvisited = set(cafe_data.keys())

    routes = []

    total_distance = 0
    total_cost = 0
    total_revenue = 0

    for _, van in vans_df.iterrows():

        capacity = van["capacity_kg"]
        cost_km = van["fuel_cost_per_km"]

        current = "depot"
        route = ["depot"]
        remaining = capacity

        while True:

            feasible = [c for c in unvisited if cafe_data[c]["weight"] <= remaining]

            if not feasible:
                break

            next_cafe = min(
                feasible,
                key=lambda x: distance_matrix.loc[current, x]
            )

            dist = distance_matrix.loc[current, next_cafe]

            total_distance += dist
            total_cost += dist * cost_km
            total_revenue += cafe_data[next_cafe]["revenue"]

            remaining -= cafe_data[next_cafe]["weight"]

            route.append(next_cafe)
            unvisited.remove(next_cafe)

            current = next_cafe

        # Return to depot when no feasible cafes can be served by van
        dist = distance_matrix.loc[current, "depot"]
        total_distance += dist
        total_cost += dist * cost_km

        route.append("depot")

        routes.append(route)

    return {
        "routes": routes,
        "total_distance": round(total_distance, 2),
        "total_revenue": round(total_revenue, 2),
        "total_cost": round(total_cost, 2),
        "profit": round(total_revenue - total_cost, 2)
    }
```

```python
# Run nearest neighbour heuristic

results = nearest_neighbour(distance_matrix, vans_df, cafe_data)

print("Routes:\n")

for i, r in enumerate(results["routes"]):
    print(f"Van {i+1}: {r}")
    print("\n")

print("Total distance:", results["total_distance"])
print("Total revenue:", results["total_revenue"])
print("Total cost:", results["total_cost"])
print("Profit:", results["profit"])
```
